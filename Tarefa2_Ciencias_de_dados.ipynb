```python
print("Olá! O objetivo desse programa é determinar se um triângulo é equilátero, isóceles ou escaleno.")
x = float(input("Para isso, você informará os lados dos 3 triângulos, a começar pelo primeiro:"))
y = float(input("Informe o valor do segundo lado:"))
z = float(input("Informe o valor do terceiro lado:"))

a = float(x + y)
b = float(y + z)
c = float(x + z)

if x < b and y < c and z < a:
  if x==y and z==x:
    print("O triângulo indicado é equilátero")

  elif x==y and x!=z or y==z and y!=x:
    print("O triângulo indicado é isóceles")

  elif x!=y and x!=z and y!=z:
    print("O triângulo indicado é escaleno")

else:
  print("Os número que indicou não obedecem a condição de existência de um triângulo, portanto, não formam um. Tente novamente.")
```
